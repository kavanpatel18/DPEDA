# Regression Plot
Seaborn's regplot() function is used to create a regression plot, which visualizes the relationship between two variables and fits a linear regression model to the data. It combines a scatter plot of the data points with a regression line and a confidence interval around that line.


```python
#Importing libraries
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Loading tips dataset
tips = sns.load_dataset('tips')
tips.head()
```

```python
#Plotting regression plot using lmplot()
sns.lmplot(x='total_bill',y='tip',data=tips)
```

```python
#Adding grid to the graph
sns.set_style('whitegrid')
sns.lmplot(x='total_bill',y='tip',data=tips)
```

```python
#Adding hue as time for the graph
sns.lmplot(x='total_bill',y='tip',data=tips,hue='time', palette='rainbow')
```

```python
#Changing the markers of the elements
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex', markers=['o','+'], palette='coolwarm')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex', markers=['o','v'], scatter_kws={'s':100} ,palette='plasma')
#sets the size of the scatter points to 100, making them larger
#scatter_kws passes a dictionary of keyword arguments to the underlying scatter plot function.
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex', markers=['o','*'], scatter_kws={'s':780} ,palette='plasma')
#sets the size of the scatter points to 780
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex', markers=['o','*'], scatter_kws={'s':5}, line_kws={'lw':10} ,palette='icefire')
# sets the line width to 10, making the regression lines much thicker
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='smoker', col='sex', row='time' ,palette='icefire')
#Each plot in the grid shows the relationship between total bill and tip, separated by whether the person is a smoker or not (hue).
#The columns of the grid are separated by sex, and the rows are separated by time.
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='smoker', col='sex', row='time', aspect=0.6 ,palette='icefire')
#aspect parameter controls the aspect ratio of each facet (individual plot) in the grid.
```

```python
res = sns.kdeplot(data=tips, x=tips['total_bill'],y=tips['tip'], fill='green')
```
