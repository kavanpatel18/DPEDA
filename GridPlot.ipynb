# Grid Plot

Grid Plot
 generally refers to a visualization where multiple plots are arranged in a grid-like structure, allowing for comparison and exploration of relationships between different variables or subsets of data.

```python
import matplotlib.pyplot as plt
import numpy as np
fig, axes = plt.subplots(1, 3, figsize=(12,4))
x = np.arange(1,11)

axes[0].plot(x, x**2, 'g--', lw=2)
axes[0].grid(True)
axes[0].set_title('Default grid')

axes[1].plot(x, np.exp(x), 'r')
axes[1].grid(color='b', ls=':', lw=0.5)
axes[1].set_title('Custom grid')

axes[2].plot(x,x)
axes[2].set_title('No grid')
fig.tight_layout() #adjusts the spacing between subplots to prevent overlapping titles or labels
plt.show()
```

```python
import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.grid('True')
ax.spines['bottom'].set_color('blue')
ax.spines['left'].set_color('red')
ax.spines['left'].set_linewidth(2)
ax.spines['right'].set_color('black')
ax.spines['top'].set_color('green')
ax.plot([5,6,7,8,1])
plt.show()
```

```python
#Using iris dataset
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
iris = sns.load_dataset('iris')
iris.head()
```

```python
#Just the grid
sns.pairplot(data=iris)
```

```python
sns.pairplot(data=iris, hue='species', palette="viridis")
```

## Pair Grids

Pair Grids takes all numberical columns and grids them up making subplots

```python
sns.PairGrid(data=iris)
```

Then you map to the grid

```python
g = sns.PairGrid(data=iris)
g.map(plt.scatter, color='magenta')
```

```python
g = sns.PairGrid(data=iris)
g.map_diag(plt.hist, color='skyblue')
g.map_upper(plt.scatter, color='salmon')
g.map_lower(sns.kdeplot)
```

```python
#Using tips dataset
tips = sns.load_dataset('tips')
tips.head()
```

```python
#Just the grid
g = sns.FacetGrid(data=tips, col='time', row='smoker')
```

```python
#The total bills are plotted as hist across the grid we created using 'map'
g = sns.FacetGrid(tips, col='time', row='smoker')
g.map(plt.hist, 'total_bill', color='green')
#Which meal do people tend to smoke?
```

```python
#Defined hue as sex and plotted a scatterplot with x axis as total bill and y axis as tip
g = sns.FacetGrid(tips, col='time', row='smoker', hue='sex', palette='dark')
g.map(plt.scatter, 'total_bill', 'tip').add_legend()
```

```python
g = sns.JointGrid(x='total_bill', y='tip', data=tips)
```

```python
g = sns.JointGrid(x='total_bill', y='tip', data=tips)
g = g.plot(sns.regplot, sns.distplot, color='black') #Regression Plot
```

```python
#Implementation of matplotlib function
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

#Dummy data
x1 = np.linspace(0.0, 5.0)
y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)

#Create 2 subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12,5))

#Plot without grid
ax1.plot(x1,y1, color='blue')
ax1.set_title('Without Grid')

#Plot with grid
ax2.plot(x1,y1, color='red')
ax2.grid(True)
ax2.set_title('With Grid')

plt.show()
```

```python
#Drawing gridlines using extra line properties like color, linestyle and linewidth
import matplotlib.pyplot as plt
import numpy as np

#dummy data
x = np.linspace(0, 2*np.pi, 400)
y = np.sin(x**2)

plt.plot(x,y, 'pink')#Set graph color

plt.title('Plot with linewidth and linestyle') #Set title

plt.grid(True, color='grey', linewidth = '0.4', ls=':', axis='y')
plt.show()
```
