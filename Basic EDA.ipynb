## Filtering Dataframes

```python
filter = stats['InternetUsers'] < 2
filter
```

```python
stats[filter]
```

```python
filter2 = stats.BirthRate > 40
stats[filter2]
```

```python
#Filtering by birthrate in one step combined by creating filter and passing it to stats brackets
stats[stats.BirthRate > 40]
```

```python
stats[filter & filter2]
```

```python
stats[(stats.BirthRate > 40)& (stats['InternetUsers'] < 2)]
```

```python
stats.InternetUsers.hist()#Generating histogram of Internet users column
```

```python
stats.hist() #Histogram of BirthRate and InternetUsers column
```

### Accessing Individual elements

.at is used for labels(even integers are treated as labels)

.iat is used for integer location

```python
stats.iat[3,4] #Accessing by integer location
```

```python
stats.iat[0,1] #Accessing by integer location
```

```python
stats.at[2,'BirthRate'] #Accessing by label
```
