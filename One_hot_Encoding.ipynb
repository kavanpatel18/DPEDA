### Creating the data

```python
import pandas as pd
```

```python
#Creating dataframe
df = pd.DataFrame({'team':['A','A','B','B','B','B','C','C'],
                  'points':[25,12,15,14,19,23,25,29]})
```

```python
df
```

Preforming one-hot encoding
### New Section

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
#Creating instance of one-hot encoder
encoder = OneHotEncoder(handle_unknown='ignore')
```

```python
#Performing one-hot encoding on team column
encoder_df = pd.DataFrame(encoder.fit_transform(df[['team']]).toarray())
```

```python
final_df = df.join(encoder_df)
```

```python
print(final_df)
```

### Drop the original categorical variable

```python
#Drop team column
final_df.drop('team',axis=1,inplace=True)
```

Practise Example

```python
#Create a dataframe with the columns as "Gender","Age","Degree"
import pandas as pd
data = {'Gender': ['Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female'],
        'Age': [25, 30, 22, 35, 28, 40, 32, 29, 26, 38],
        'Degree': ['BSc', 'MSc', 'PhD', 'BSc', 'MSc', 'PhD', 'BSc', 'MSc', 'PhD', 'BSc']}
df = pd.DataFrame(data)
print(df)
```

```python
# Apply one hot encoding on the column gender
encoder_gender = OneHotEncoder(handle_unknown='ignore')
encoder_gender_data = encoder_gender.fit_transform(df[['Gender']]).toarray()
encoder_gender_df = pd.DataFrame(encoder_gender_data, columns=encoder_gender.get_feature_names_out(['Gender']))

# Apply one hot encoding on the column degree
encoder_degree = OneHotEncoder(handle_unknown='ignore')
encoder_degree_data = encoder_degree.fit_transform(df[['Degree']]).toarray()
encoder_degree_df = pd.DataFrame(encoder_degree_data, columns=encoder_degree.get_feature_names_out(['Degree']))

# Concatenate the encoded dataframes
encoded_df = pd.concat([encoder_gender_df, encoder_degree_df], axis=1)

# Join the concatenated dataframe with the original dataframe, dropping the original columns
final_df = df.drop(['Gender', 'Degree'], axis=1).join(encoded_df)

# Display the final dataframe
display(final_df)
```
