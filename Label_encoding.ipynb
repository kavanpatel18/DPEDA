In Python, label encoding is a technique for converting categorical data (like text) into a numerical format that machine learning algorithms can understand. It assigns a unique integer to each distinct category in a dataset, which is a crucial data preprocessing step. LabelEncoder assigns an integer value to each unique category it finds, based on alphabetical order.

```python
#Importing required libraries
import pandas as pd
import numpy as np
```

```python
#Creating initial dataframe
bridge_types = ('Arch','Beam','Truss','Cantilever','Tied Arch','Suspension')
bridge_df = pd.DataFrame(bridge_types, columns=['Bridge_Types'])
bridge_df
```

```python
#Converting types of columns to 'category'
bridge_df['Bridge_Types'] = bridge_df['Bridge_Types'].astype('category')
```

```python
#Assigning numberical values and storing in another column
bridge_df['Bridge_Types_Cat'] = bridge_df['Bridge_Types'].cat.codes
bridge_df
```

### Using scikit learn library to encode

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
```

```python
labelencoder = LabelEncoder()
```

```python
bridge_df["Bridge_Types_Cat"] = labelencoder.fit_transform(bridge_df["Bridge_Types"])
bridge_df
```

```python
from sklearn.preprocessing import OneHotEncoder
enc = OneHotEncoder(handle_unknown='ignore')
enc_df = pd.DataFrame(enc.fit_transform(bridge_df[['Bridge_Types_Cat']]).toarray())
bridge_df = bridge_df.join(enc_df)
bridge_df
```
