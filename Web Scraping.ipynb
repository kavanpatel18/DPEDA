## Scraping a website using Beautiful Soup

```python
from bs4 import BeautifulSoup
```

```python
import requests
```

```python
url = "https://www.sitpune.edu.in"
```

```python
response = requests.get(url)
```

```python
print(response)
```

```python
html_content = response.content
```

```python
soup = BeautifulSoup(html_content, "html.parser")
```

## Extracting the title

```python
soup.find("title")
```

```python
soup.find("title").string
```

## Extracting the h1 tags

```python
h1_tags = soup.find_all("h1")
```

```python
h1_tags
```

```python
for h1 in h1_tags:
    print(h1.text)
```

## Extracting h2 tags

```python
h2_tags = soup.find_all("h2")
```

```python
h2_tags
```

```python
for h2 in h2_tags:
    print(h2.text)
```

## Extracting h3 tags

```python
h3_tags = soup.find_all("h3")
```

```python
for h3 in h3_tags:
    print(h3.text)
```

## Extracting the < a > tags


```python
a_tags = soup.find_all("a")
```

```python
len(a_tags)
```

```python
a_tags
```

## Extracting the links from the < a > tags

```python
list_of_links = []
for a in a_tags:
    link = a.get("href")
    if link != None:
        list_of_links.append(link)
```

```python
len(list_of_links)
```

```python
list_of_links
```

## Making a dataframe of the links

```python
import pandas as pd
```

```python
df_links = pd.DataFrame(data=list_of_links, columns=["Link"])
```

```python
df_links.shape
```

```python
df_links
```

## Saving it as a .csv file

## Extracting all the images from the website

```python
img_links = []
```

```python
link = soup.select('img[src^="https://www.sitpune.edu.in/assets/images"]')
```

```python
for img in link:
    img_links.append(img['src'])
```

```python
directory = '/content/Scrap_images'
```

```python
for index, img_link in enumerate(img_links):
    img_data = requests.get(img_link).content
    with open(f'{directory}/'+str(index+1)+'.jpg','wb+') as f:
        f.write(img_data)
```
