# **KNN imputer for quantitative and qualititave data**

Import necessary libraries

```python
import numpy as np
import pandas as pd
```

Import the KNNimputer

```python
from sklearn.impute import KNNImputer
#This class is used for imputing missing values in a dataset using the k-nearest neighbors (KNN) algorithm. It can handle both quantitative and qualitative data.
```

Create dataset for marks of students


```python
dict = {'Maths':[85, 91, np.nan, 91],
        'Chemistry':[72, 75, 74, np.nan],
        'Physics':[np.nan, 68, 82, 81],
        'Biology':[81,81,72,np.nan]
}
```

Create a dataframe from the Dictionary

```python
before_imputation = pd.DataFrame(dict)
before_imputation
```

Replace the missing values with KNNimputer

```python
imputer = KNNImputer(n_neighbors=10)
after_imputation=imputer.fit_transform(before_imputation)
after_imputation
#applies the imputer to the before_imputation DataFrame using fit_transform, which calculates and replaces the missing values based on the 10 nearest neighbors.
```

Converting the array into dataframe

```python
before_imputation.columns
```

```python
after_imputation = pd.DataFrame(after_imputation, columns=before_imputation.columns)
after_imputation
#converts the after_imputation NumPy array into a pandas DataFrame.
#It uses the column names from the original before_imputation DataFrame to ensure the new DataFrame has the correct labels for each column
```

## import Libraries

```python
import pandas as pd
import numpy as np
from sklearn.impute import KNNImputer
```

## Create dataframe



```python
data = [[1, 3, np.nan, 4], [6, np.nan, 8, np.nan], [5, 4, 2, 3], [9, np.nan, 6, 8]]
column_names = ['Class A', 'Class B', 'Class C', 'Class D']
X = pd.DataFrame(data, columns=column_names)
display(X)
```

## Analyze missing values




```python
missing_values_per_column = X.isnull().sum()
total_missing_values = missing_values_per_column.sum()
percentage_missing_values = (total_missing_values / X.size) * 100

print("Number of missing values per column:")
print(missing_values_per_column)
print("\nTotal number of missing values in the DataFrame:")
print(total_missing_values)
print("\nPercentage of missing values in the DataFrame:")
print(percentage_missing_values)
```

## Handle missing values by dropping


```python
X_dropped_rows = X.dropna(axis=0)
display(X_dropped_rows)

X_dropped_columns = X.dropna(axis=1)
display(X_dropped_columns)
```

## Handle missing values by filling



```python
column_means = X.mean()
X_filled_mean = X.fillna(column_means)
display(X_filled_mean)
```

## Impute missing values using knn



```python
imputer = KNNImputer(n_neighbors=2)
after_imputation = imputer.fit_transform(X)
Y = pd.DataFrame(after_imputation, columns=X.columns)
display(Y)
```
