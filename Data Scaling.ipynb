# **Data Scaling**

Data scaling is a crucial preprocessing step in machine learning that involves transforming the values of numerical features in a dataset to a common scale. This technique helps ensure that no single feature dominates the learning process due to its larger magnitude or variance, thereby improving the accuracy and stability of machine learning models

Common data scaling techniques
1. Normalization (Min-Max Scaling): Rescales features to a fixed range, typically 0 to 1. It's useful when the data distribution is unknown but can be sensitive to outliers.  
2. Standardization (Z-score Scaling): Transforms features to have a mean of 0 and a standard deviation of 1. It's suitable for normally distributed data or when the distribution is unknown and is less affected by outliers than normalization.




```python
import pandas as pd
import numpy as np
#import required libraries
```

## **1.** **Min_Max Scaling (Normalization)**



**Import Min-Max Scaler (Normalization Scaler)**

```python
from sklearn.preprocessing import MinMaxScaler
```

**Define Data**

```python
data = pd.DataFrame(data = [[5,9],[10,2],[2,0],[0,10]], columns=['X','Y'])
```

```python
data
```

**Plot scatter plot**

```python
import matplotlib.pyplot as plt
plt.scatter(data.X, data.Y)
plt.axvline(x=0, c="red")
plt.axhline(y=0, c="green")
plt.show()
```

**Define, fit and tranform min-max scaler**

```python
scaler = MinMaxScaler()
min_max_scaled_data = scaler.fit_transform(data)
```

**Print Scaled data**

```python
min_max_scaled_data = pd.DataFrame(min_max_scaled_data, columns=['X','Y'])
min_max_scaled_data
```

```python
import matplotlib.pyplot as plt
plt.scatter(min_max_scaled_data.X, min_max_scaled_data.Y)
plt.axvline(x=0, c="red")
plt.axhline(y=0, c="green")
plt.show()
```

## **Subplot**

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 5))

plt.subplot(1,2,1)
plt.scatter(data.X, data.Y)
plt.axvline(x=0, c="red")
plt.axhline(y=0, c="green")

plt.subplot(1,2,2)
plt.scatter(min_max_scaled_data.X, min_max_scaled_data.Y)
plt.show
```

## **2.** **Z-Score Scaling (Standardization Scaler)**



```python
from sklearn.preprocessing import StandardScaler
```

```python
data = pd.DataFrame(data=[[5,5],[10,2],[2,0],[0,10]], columns = ['X','Y'])
data
```

```python
scaler1 = StandardScaler()
std_scaled = scaler1.fit_transform(data)
std_scaled
```

```python
std_scaled = pd.DataFrame(std_scaled, columns=['X','Y'])
std_scaled
```

```python
std_scaled.describe()
```

**Plot Standard Scaler**

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(5,5))
plt.scatter(std_scaled.X, std_scaled.Y)
plt.axvline(x=0, c="brown")
plt.axhline(y=0, c="green")
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(15, 5))
plt.subplot(1,2,1)
plt.scatter(data.X, data.Y)
plt.axvline(x=0, c="red")
plt.axhline(y=0, c="green", label='y=0')

plt.subplot(1,2,2)
plt.scatter(std_scaled.X, std_scaled.Y)
plt.axvline(x=0, c="brown")
plt.axhline(y=0, c="green")
plt.show()
```

## Plot both min_max and standard scaler

```python
import matplotlib.pyplot as plt
plt.scatter(std_scaled.X, std_scaled.Y, marker= '^')
plt.scatter(min_max_scaled_data.X, min_max_scaled_data.Y, marker= 'o')
plt.axvline(x=0, c="brown")
plt.axhline(y=0, c="green")
plt.show()
```
