```python
import vertexai
from vertexai.preview.vision_models import ImageGenerationModel
from vertexai.generative_models import GenerativeModel, Part

# Set your project and location
project_id = "qwiklabs-gcp-01-68287294602a"
location = "us-central1"
output_file = "image.jpeg"

def generate_bouquet_image(prompt: str) -> None:
    """Generates a bouquet image from a text prompt and saves it locally."""
    vertexai.init(project=project_id, location=location)
    model = ImageGenerationModel.from_pretrained("imagen-3.0-generate-002")
    images = model.generate_images(
        prompt=prompt,
        number_of_images=1,
        seed=1,
        add_watermark=False,
    )
    images[0].save(location=output_file)
    print(" Image generated and saved as 'image.jpeg'.")

def analyze_bouquet_image(image_path: str) -> None:
    """Analyzes the saved bouquet image and generates a birthday wish."""
    vertexai.init(project=project_id, location=location)
    model = GenerativeModel("gemini-2.0-flash-001")

    with open(image_path, "rb") as img_file:
        image_data = img_file.read()

    image_part = Part.from_data(data=image_data, mime_type="image/jpeg")

    print(" Generating birthday wish based on the bouquet image...\n")
    stream = model.generate_content(
        [
            image_part,
            "Based on this bouquet, generate a beautiful birthday wish."
        ],
        stream=True
    )

    for chunk in stream:
        print(chunk.text, end="")

# Main flow
if __name__ == "__main__":
    generate_bouquet_image("Create an image containing a bouquet of 2 sunflowers and 3 roses")
    analyze_bouquet_image(output_file)
```
