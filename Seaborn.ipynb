# **Seaborn**

Seaborn is a powerful and popular open-source Python library for creating high-level, informative, and visually appealing statistical graphics.

```python
import seaborn as sns #import seaborn

sns.set_theme(style="darkgrid")
```

```python
tips = sns.load_dataset("tips") #Load an example dataset
tips.head()
```

```python
sns.barplot(x="sex", y="total_bill", data=tips)
```

```python
import numpy as np
```

## countplot

A count plot in Seaborn, implemented using the seaborn.countplot() function, is a specialized type of bar plot designed to display the count of observations within each category of a categorical variable. It visually represents the distribution of a categorical variable by showing the number of times each unique value appears in the dataset.

```python
sns.countplot(x="day", data=tips)
```

## Boxplot
A box plot, also known as a box-and-whisker plot, is a graphical representation of the distribution of numerical data through their quartiles. Seaborn, a Python data visualization library built on Matplotlib, provides the seaborn.boxplot() function to easily create these plots.

```python
sns.boxplot(x='day', y='total_bill', data=tips, palette='rainbow')
```

```python
sns.boxplot(data=tips, palette='rainbow', orient='h')
#Entire dataframe and changing the orientation
```

```python
sns.boxplot(x='day', y='total_bill', hue='sex', data=tips, palette='magma')
#hue parameter is used to map a categorical variable to the color of the plot elements
```

## violinplot
A violin plot plays a similar role as a box-and-whisker plot. It shows the distribution of data points after grouping by one (or more) variables. Unlike a box plot, each violin is drawn using a kernel density estimate of the underlying distribution.



```python
sns.violinplot(x='day', y='total_bill', data=tips, palette='viridis')
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='sex' ,palette='Set1')
#hue parameter determines which column in the data frame should be used for colour encoding
```

## Displot
This function provides access to several approaches for visualizing the univariate or bivariate distribution of data, including subsets of data defined by semantic mapping and faceting across multiple subplots.

```python
sns.displot(data=tips, x='total_bill', col='time', kde=True)
```

```python
sns.distplot(tips['total_bill']) #older version of displot
```

```python
sns.distplot(tips['total_bill'], kde=True, bins=5)
#"bin size" refers to the width of the intervals used to group numerical data
```

```python
#To remove the kde layer and just have the histogram
sns.distplot(tips['total_bill'], kde=True, bins=30)
```

## Jointplot
Draw a plot of two variables with bivariate and univariate graphs.
The seaborn.jointplot() function in Python is a powerful and convenient tool for visualizing the relationship between two variables while also displaying their individual distributions.

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='scatter', color='navy')
#kind='scatter' creates a scatter plot in the main part of the figure, showing each individual data point.
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex', color='magenta')
#compute a bivariate histogram using hexagonal bins
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='reg', color='brown')
#add a linear regression fit (using regplot()) and univariate KDE curves
```

## Pairplot
Plot pairwise relationships in a dataset.

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips, hue='sex', palette='magma')
```

## Rugplot
Plot marginal distributions by drawing ticks along the x and y axes.

This function is intended to complement other plots by showing the location of individual observations in an unobtrusive way.

```python
sns.rugplot(tips['day'])
```

These two plots visually represent the distribution of tip amounts in the dataset, with the KDE plot showing the overall shape of the distribution and the rug plot showing the exact values of each tip

```python
sns.kdeplot(tips['tip'])
sns.rugplot(tips['tip'])
```
