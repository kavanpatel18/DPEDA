### ax.bar (x, height, width, bottom, align)

```python
import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
langs = ['C','C++','Java','Python','PHP']
students = [23,17,35,29,12]
ax.bar(langs,students)
plt.show()
```

### Plotting multiple graphs

```python
import numpy as np
import matplotlib.pyplot as plt
data = [[30, 25,50,20],[40,23,51,17],[35,22,45,19]]
x = np.arange(4)
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.bar(x + 0.00, data[0], color = 'pink', width = 0.25)
ax.bar(x + 0.25, data[1], color = 'g', width = 0.25)
ax.bar(x + 0.50, data[2], color = 'r', width = 0.25)
plt.show()
```

### Create a stacked bar chart

```python
import numpy as np
import matplotlib.pyplot as plt
n = 5
meanMenscore = (20, 35, 30, 35, 27)
meanWomenScore = (32, 35, 30, 35, 27)
ind = np.arange(n) #The x locations for the groups
width = 0.35
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.bar(ind, meanMenscore, width, color = 'pink')
ax.bar(ind, meanWomenScore, width, bottom=meanMenscore, color = 'CYAN')
ax.set_ylabel('Scores') #sets the label for the y-axis.
ax.set_title('Scores by group and gender') #sets the title of the chart
ax.set_xticks(ind)
ax.set_xticklabels(('G1', 'G2', 'G3', 'G4', 'G5'))
ax.set_yticks(np.arange(0,81,10)) #sets the positions of the ticks on the y-axis, starting from 0, going up to 80, with a step of 10
ax.legend(('Men', 'Women'))
plt.show()
```

## Seed Function
