#Data Wrangling


Data wrangling, also known as data munging, is the process of transforming and mapping data from its raw form into a more suitable format for analysis, storage, and machine learning. It involves cleaning, structuring, and transforming data to improve its quality and usability.

The typical data wrangling process includes the following steps:


1.   Structuring: Organizing data into a format suitable for analysis, such as tabular form.
2.   Cleaning: Identifying and correcting errors, missing values, and duplicates. This involves techniques like removing duplicates and handling missing data.
3.    Enriching: Adding new information or merging data with external sources.
Validating: Verifying the accuracy and quality of the processed data.
4.    Publishing: Preparing the final dataset for analysis or dissemination. Documentation of the process is also important.
5. Discovery: Understanding the data's source, structure, and potential issues.
6. Validating: Verifying the accuracy and quality of the processed data.

## Handling missing values

```python
df.info() #provides a concise summary of the DataFrame
```

```python
df.isnull().any()#tells you which columns have at least one missing value.
```

```python
df.isna().sum() #displays the total number of missing values in each column
```

```python
df.isna().sum().sum() #Total missing values
```

```python
#Sort in descending order
df.isna().sum().sort_values(ascending=False)
```

```python
(df.isnull().sum() / df.shape[0] * 100).sort_values(ascending=False)
#Percentage of null values in different rows
#df.shape[0] divides the number of missing values in each column by the total number of rows
```

```python
#Filling missing values in the age column
age_mean_value = df['Age'].mean()
df['Age'] = df['Age'].fillna(age_mean_value)
df
```

```python
df.isna().sum()
#Check there are no null values in age column
```

```python
df['Embarked'].fillna(method='ffill', inplace=True)
df.isna().sum()
#Forward fill propagates the last valid observation forward to fill the missing values.
#For example, if a row has a missing 'Embarked' value, it will be filled with the 'Embarked' value from the row directly above it that is not missing.
```

Creating a new Title column

```python
df['Title'] = df['Name'].str.extract('([A-Za-z]+)\.', expand=True) #Creating new Title column
df['Title'].value_counts()
```

Replace rare titles with more common ones

```python
mapping = {'Mlle': 'Miss', 'Major': 'Mr', 'Col' : 'Mr', 'Sir':'Mr', 'Don':'Mr', 'Mme':'Mr', 'Jonkheer':'Mr', 'Lady':'Mrs', 'Capt':'Mr', 'Countess':'Mrs', 'Ms':'Miss', 'Dona':'Mrs'}
df.replace({'Title': mapping}, inplace=True)
```

```python
df['Title'].value_counts()
```

```python
df.groupby('Title')['Age'].median()
```

Impute missing age values using median of Title groups

```python
title_ages = dict(df.groupby('Title')['Age'].median())
title_ages
```

Create a column of the average ages

```python
df['age_med'] = df['Title'].apply(lambda x: title_ages[x])
df
```
