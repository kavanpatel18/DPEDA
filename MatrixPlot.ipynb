# **Matrix Plot**
Creating a matrix plot in Python typically refers to visualizing a 2D array or matrix where the values are represented by colors.

```python
#Importing modules
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
#Generating 2D 10x10 matrix of random numbers from 1 to 100
data = np.random.randint(low=1, high=100, size=(10,10))
print("The data to be plotted:\n",data)
```

```python
#Plotting the heatmap
hm = sns.heatmap(data, cmap="YlGnBu")
plt.show()
```

```python
#Printing the values on the graph
data = np.random.randint(low=1, high=100, size=(10,10))
hm = sns.heatmap(data, cmap="YlGnBu", annot=True)
plt.show()
```

```python
#Changing the theme of the graph by colormap
data = np.random.randint(low=1, high=100, size=(10,10))
hm = sns.heatmap(data, cmap="tab20", annot=True)
```

```python
#Setting linewidth and linecolor for the graph
data = np.random.randint(low=1, high=100, size=(10,10))
hm = sns.heatmap(data, linewidth = 2, linecolor='yellow')
```

```python
#Removing the colorbar of the graph
data = np.random.randint(low=1, high=100, size=(10,10))
hm = sns.heatmap(data, cbar=False)
```

```python
#Removing the labels of x and y axis of the graph
data = np.random.randint(low=1, high=100, size=(10,10))
hm = sns.heatmap(data, xticklabels=False, yticklabels=False, cmap='cool')
```

```python
tips=sns.load_dataset('tips')
tips.head()
```

```python
#Matrix form for correlation data
tips.corr(numeric_only=True)
```

```python
#Plotting the heatmap
sns.heatmap(tips.corr(numeric_only=True), cmap="YlGnBu")
```

```python
sns.heatmap(tips.corr(numeric_only=True), cmap="YlGnBu", annot=True)
```

```python
flights = sns.load_dataset('flights') #Loading flights dataset
flights.head()
```

```python
flights.pivot_table(values='passengers', index='month', columns='year')
#creates a matrix where each cell shows the number of passengers for a specific month and year.
```

```python
pvflights = flights.pivot_table(values='passengers', index='month', columns='year')
sns.heatmap(pvflights, cmap='YlGnBu')
#Plotting the pivottable into a heatmap
```

```python
sns.heatmap(pvflights, cmap='magma', linecolor='white', linewidths=1)
```
